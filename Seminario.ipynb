# INTRODUCCION Y CONTEXTO


El RMS Titanic fue un transatlántico británico cuya construcción finalizó en 1912, año en el que se convirtió tanto en el mayor barco de pasajeros del mundo como en el objeto móvil más grande jamás creado. El Titanic era un compendio de lujos al que se le dio también una publicidad que ensalzaba su seguridad, llegando a ser presentado como el “insumergible”. El barco zarpó de Southampton (Inglaterra), hizo escala en Cherburgo (Francia) y en Queenstown (Irlanda), y posteriormente tenía como destino la ciudad de Nueva York.  El naufragio del Titanic fue una catástrofe marítima ocurrida en la noche del 14 al 15 de abril de 1912, cuando chocó contra un iceberg en el océano Atlántico frente a la costa de Terranova, lo que provocó el hundimiento del barco en menos de tres horas, a las 02:20 del 15 de abril. Había alrededor de 2227 personas a bordo del Titanic, aunque es imposible determinar el número exacto debido a cancelaciones de última hora y a la presencia de pasajeros clandestinos, y la misma inexactitud ocurre con el número de fallecidos, que fue de unas 1514 personas, lo que convirtió a este naufragio en el más fatal de la época y en una de las mayores tragedias marítimas en tiempos de paz.

El archivo “train.csv” contiene los detalles de un
subconjunto de 891 pasajeros a bordo del Titanic y también revela si sobrevivieron
o no al naufragio.
Del conjunto de 12 variables iniciales de las que disponíamos en el archivo (contando con la variable objetivo, que es la supervivencia) se fueron descartando por diversos motivos algunas de ellas,
principalmente por no considerarlas de gran relevancia de cara a explicar dicha supervivencia al naufragio, y también por falta de datos o por correlación elevada
con algún otro atributo considerado. Otras variables se transformaron, reduciendo finalmente las características destacables a 7 (3 de ellas numéricas y 4 categóricas), que son las siguientes:

* Survived: Indica si el pasajero sobrevivió o no. Tiene dos posibles valores:
 - 0: El pasajero murió.
 - 1 : El pasajero sobrevivió.
* Pclass : Clase en la que viajaba el pasajero. Tiene tres posibles valores:
  -1: Pasajeros que viajaban en primera clase
  -2: Pasajeros que viajaban en segunda clase
  -3: Pasajeros que viajaban en tercera clase
* Sex: Género del pasajero.
* Age: Edad del pasajero.
* Fare: Indica la tarifa abonada por el ticket del barco.
* Embarked: Indica la ciudad donde embarcó el pasajero. Tiene tres posibles valores:
  - S: El pasajero embarcó en el puerto de Southampton, Inglaterra.
  - C: El pasajero embarcó en el puerto de Cherbourg, Francia.
  - Q: EL pasajero embarcó en el puerto de Queenstown, Irlanda.
* Num_Familiares: Número de familiares de un pasajero que iban en el barco.

A su vez, dos de estas variables, Age y Num_Familiares, se han utilizado de distintas formas en lo que a la realización del trabajo se refiere. La variable Age, que es continua, en ocasiones se ha discretizado por intervalos (dando lugar a la variable GrupoEdad), y la variable Num_Familiares, que es numérica, en ocasiones se ha utilizado como categórica para distinguir simplemente si el pasajero llevaba familia a bordo o no (dando lugar a la variable Family)

La variable Fare la redondeo a un número de 2 decimales, y la variable Age la convierto en un número entero (según se indica en la descripción del dataset, la edad aparece como fraccionaria si es menor que 1 o si se estimó la edad de algún
pasajero).
Debido a la existencia de missing values en las variables Embarked y Age también tendremos que procesar los datos en ese sentido. En el caso de la
variable Embarked tan sólo son 2 observaciones con valor “NA”, sutituyéndose éste por el valor “S”, correspondiente a la moda de dicha variable. El

#Interpretación y conclusiones

En este trabajo se ha llevado a cabo, a través de distintas visualizaciones de datos, el análisis del dataset del Titanic que kaggle ha puesto a nuestra disposición. Se han mostrado la distribución de los datos, las relaciones más importantes entre las variables , y también de qué manera y en qué proporción influían éstas a la hora de explicar la variable objetivo, que es clasificar a un pasajero como víctima o superviviente de la tragedia. En este sentido se ha visto que algunas características como pagar un precio más elevado por el billete, ir en primera clase, embarcar en el puerto de Cherbourg, o tener familia a bordo, elevaban las posibilidades de sobrevivir. En el lado opuesto, un precio bajo por el billete, ir en tercera en clase, haber embarcado en Queenstown o viajar solo, te acercaban a la muerte. 

Mención aparte merecen las variables edad y género del pasajero. Ambas se han analizado desde el punto de vista de comprobar si el caso del Titanic puede tomarse como ejemplo de una popular ley no escrita en el mar, "las mujeres y los niños primero", y se ha comprobado que así es. Particularmente llamativo es el caso de las mujeres, que disfrutaron de grandes ventajas a la hora de ponerse a salvo respecto a los hombres, ya que sobrevivieron aproximadamente 3 de cada 4 mujeres por tan solo uno de cada 5 hombres. En cuanto al tema de los niños, se observó como fueron el único grupo de edad con más probabilidades de vivir que de morir, esto solo en lo que respecta a los menores de 12 años, que es la edad hasta la que se hacía dicha consideración en la época. 

Por lo tanto, se puede afirmar sin miedo a equivocarse, que la evacuación y salvamento del Titanic se basó por un lado en directrices económicas y de clase social, lo cual podría considerarse un motivo clásico y reiterado a lo largo del tiempo para cualquier ámbito, y por otro lado en el imperativo categórico que anteponía la salvación de mujeres y niños. En este caso, esta manera de actuar sí que resulta novedosa, ya que por mucho que pueda existir la creencia de que esta manera de proceder es la que se llevará a cabo en situaciones límite como la del Titanic, la realidad es que esto es solo un mito, y los recientes estudios realizados por investigadores de la Universidad de Uppsala (el cual aparece publicado en la revista Proceedings de la Academia Nacional de Ciencias(PNAS)), en Suecia, así lo demuestran. En este estudio, los investigadores analizaron datos sobre 18 grandes desastres marítimos desde 1852 a 2011, lo que incluye a 15.000 pasajeros y miembros de la tripulación de 30 nacionalidades diferentes.Las conclusiones extraídas en lo que a salvamento se refiere, fueron diametralmente opuestas a las del Titanic, ya que en la gran mayoría de estos naufragios se salvaron muchos más hombres que mujeres y niños. De hecho, parece ser que el Titanic fue la última gran excepción y uno de los grandes ejemplos en la aplicación de dicho imperativo, en cuyo cumplimiento tuvo mucho que ver el capitán del barco, que fue quién ordenó que las mujeres y niños fueran puestos a salvo los primeros, e incluso existen informes de que algunos oficiales dispararon a los pasajeros varones que intentaban desobedecer la orden y subirse a los botes. Por lo tanto, abstrayéndonos un poco de lo que nos cuentan en sí mismos los datos, podríamos dictaminar que en este caso del Titanic, la variable más importante y la que explica más en lo que al salvamento de los pasajeros se refiere, es el capitán.

Incidir pues, que tal y como se indica en el estudio mencionado, esta práctica del "sálvase quién pueda" es la que parece predominar. De igual forma, también suelen salvarse más personas de la tripulación que del pasaje, capitán incluido, hecho que tampoco se cumplió en el Titanic. Con dichos datos en la mano, se pretende invitar a la reflexión sobre la normalización/regulación de esta circunstancia, ya que si bien es verdad que hoy en día existen los medios a priori para salvar a todas las personas que van en un barco, cualquier eventualidad puede hacer que no sea así, y hay casos recientes que lo confirman.
